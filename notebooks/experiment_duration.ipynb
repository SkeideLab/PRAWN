```python
import mne
import json
import pandas as pd
```

```python
with open('../sessions.json', "r") as f:
    sessions_file = json.load(f)
    sessions = sessions_file['allsessions_prediction']

sessions
    
```

```python
df = []

for session in sessions:
    raw = mne.io.read_raw_fif(f'../data/interim/{session}/raw.fif', preload=True)
    
    # Get the number of time points and sampling frequency
    n_times = raw.n_times  # Total number of samples
    sfreq = raw.info['sfreq']  # Sampling frequency in Hz (samples per second)

    # Calculate the duration in seconds
    duration_minutes = n_times / sfreq / 60.0
    
    sub = session.split('_')[0]
    ses = session.split('_')[1]
    
    df = df + [{'subject': sub, 'session': ses, 'duration_minutes': duration_minutes}]

    
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
# mean session duration (incl. breaks)
print(df['duration_minutes'].mean())

# min and max session duration
df['duration_minutes'].min(), df['duration_minutes'].max()
```

```python
# mean session duration (sum of both session) per subject
df.groupby('subject')['duration_minutes'].sum().mean()
```
